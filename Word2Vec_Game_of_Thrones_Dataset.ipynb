```python
import gensim
import os
```

```python
import nltk
nltk.download('punkt_tab')
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('/content/data'):
    try:
        with open(os.path.join('data', filename), 'r', encoding='latin-1') as f:
            corpus = f.read()
            raw_sent = sent_tokenize(corpus)
            for sent in raw_sent:
                story.append(simple_preprocess(sent))
        print(f"Decoded file: {filename}")
    except UnicodeDecodeError:
        print(f"Could not decode file: {filename}. Skipping.")
    except Exception as e:
        print(f"An error occurred while processing file: {filename} - {e}")
```

```python
story
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

story_filtered = []
for sentence in story:
    filtered_sentence = [word for word in sentence if word not in stop_words]
    story_filtered.append(filtered_sentence)

story = story_filtered
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
